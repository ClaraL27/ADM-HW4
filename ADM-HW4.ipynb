# 2. Grouping songs together!

```python
import pandas as pd
from tqdm.notebook import tqdm
```

## 2.1 Getting your data!

```python
echonest = pd.read_csv('csv/echonest.csv')
features = pd.read_csv('csv/features.csv')
tracks = pd.read_csv('csv/tracks.csv')
```

```python
#num rows echonest
N_ECHONEST = 13129

extract_tracks = pd.DataFrame()
extract_features = pd.DataFrame()

for index, row in tqdm(echonest.iterrows(), total=N_ECHONEST):
    track_id = row['track_id']

    matching_track = pd.DataFrame(tracks[tracks['track_id'] == track_id])
    matching_feature = pd.DataFrame(features[features['track_id'] == track_id])

    extract_tracks = extract_tracks.append(pd.DataFrame(matching_track))
    extract_features = extract_features.append(pd.DataFrame(matching_feature))
    
    #DEBUG
    if index == 1000:
        break

```

```python
#merge
new_df = extract_tracks.merge(echonest)
main_df = new_df.merge(extract_features)
```
