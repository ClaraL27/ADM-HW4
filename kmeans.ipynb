```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

import kmeans

from importlib import reload
reload(kmeans)

X, y = make_blobs(
    n_samples=500,
    n_features=2,
    centers=4,
    cluster_std=1,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=1,
)

data = pd.DataFrame(X)

inertia = kmeans.elbow(data, range(2,10), plot = True)
avg_silh = kmeans.avg_silhouette(data, range(2,10), plot = True)
```

```python
model = kmeans.KMeans(K = 4)
model.fit(data)

silh = model.silhouette(data, plot = True)
```
