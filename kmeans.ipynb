```python
from sklearn.datasets import make_blobs

X, y = make_blobs(
    n_samples=500,
    n_features=2,
    centers=4,
    cluster_std=1,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=1,
)

data = pd.DataFrame(X)

model = kmeans.KMeans(K = 5)
model.fit(data)

model.display(X[:,0], X[:,1], 
              show_centroids = True)
```
